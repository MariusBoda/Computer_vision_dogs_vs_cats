```python
import os
import numpy as np
from PIL import Image

# Step 1: Load Images and Assign Labels
def load_images_from_folder(folder, label):
    images = []
    labels = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        img = Image.open(img_path).convert('L')  # Convert to grayscale for simplicity
        img = img.resize((64, 64))  # Resize images to a fixed size (e.g., 64x64)
        img_array = np.array(img).flatten()  # Flatten the image
        images.append(img_array)
        labels.append(label)
    return images, labels

# Paths to the folders with images
folder_path = '/kaggle/input/house-plant-species/house_plant_species'
categories = os.listdir(folder_path)  # Assumes each folder is a category
X, y = [], []

# Load all images and labels
for idx, category in enumerate(categories):
    category_folder = os.path.join(folder_path, category)
    images, labels = load_images_from_folder(category_folder, idx)
    X.extend(images)
    y.extend(labels)
```
